# Ejercicio 2: Preguntas

## ¿Qué es Jupyter y cómo puede serle útil a un científico de datos?

Jupyter es un entorno de programación interactivo, que se asemeja a los entornos integrados de desarrollo.
El proceso consiste en crear documentos, llamados notebooks, cuya unidad basica son celdas donde se escribe codigo. Este codigo puede ser ejecutado, como se haria en una terminal, y el resultado se observa directamente debajo de la celda.

Es de utilidad a los desarrolladores e investigadores debido al formato interactivo, ya que tradicionalmente se ha hecho con la terminal solamente, lo cual puede confundir y hacer el proceso mas lento para personas menos familiarizadas con el mundo de la computación. En el caso especifico de un cientifico de datos, es de vital importancia hacer prototipos de algorimos para analizar y transformar datos rapidamente; y tambien es esencial visualizar los resultados graficamente. El ambiente interactivo de jupyter es bastante adecuado para hacer realizar dinamicamente estas actividades.

## ¿Qué es una celda markdown?

Son celdas que especificamente muestran el resultado de un documento markdown. El markdown es un lenguaje de creacion de documentos,
muy extendido a traves de la industria tecnologica, que permite escribir anotaciones mediante la combinacion de texto y caracteres especiales del lenguaje, y que posteriormente sean interpretados y mostrados como un documento presentable.

Se permiten construir elementos como titulos, tablas, listas, anotaciones, etc. con el uso de simbolos como #,*,-; sin necesidad de usar un procesador de texto para agregar el estilo. Se guardan en archivos de texto plano, lo cual facilita su almacenaje y envio a traves de la web.

## ¿Qué es un paquete (o librería) en Python?

Los paquetes en python son la unidad mas grande para compartir o reutilizar codigo en el lenguaje.
Consisten de directorios, que contienen archivos en python que definen clases, funciones y constantes de utilidad.
Los paquetes normalmente se crean para cumplir una tarea o series de tareas relacionadas: hay paquetes para el procesado de imagenes, analisis financieros y estadiscos, manejo de graficos, etc.

## Describa para qué sirven cada una de las siguientes librerías: 

#### pip: 
> Es el administrador de paquetes nativo de python; una herramienta CLI encargada de instalar, desinstalar y actualizar paquetes de python en la maquina.

### numpy: 
> Es una biblioteca que contiene clases para representar conceptos del algebra lineal y el análisis numerico: contiene clases para representar vectores, matrices, tensores, etc. y funciones que para operar sobre estos.

### pandas:
> Es uno de los paquetes estandard para la manipulación y el análisis de datos. Contiene herramientas para representar tablas, series datos, funciones para calcular métricas estadisticas y aplicar transformaciones sobre datos.

### seaborn: 
> Paquete de graficación estadística, basada en matplotib. Se enfoca en la creación simplificada de graficos estadisticos, ya sean histogramas, graficos de barra, caja, pastel, etc.

### matplotlib: 
> Paquete general de graficación de python. Tiene la capacidad de construir gráficas, animaciones, visualizaciones, etc. por medio de codigo, y poder mostrarlos en la maquina, o guardarlo en diferentes formatos.

### scikitlearn:

> Paquete standard de python para el aprendizaje automático. Continene interfaces para limpiar y transformar datos; y entrenar algoritmos de aprendizaje supervisado y no supervisado. 

### requests: 
> Paquete que permite usar de forma simplificada el protocolo HTTP, para poder recibir y enviar información a traves de la web, por medio de Constructos. Esto nos interesa para poder obtener información de paginas web, documentos, datos estadisticos, etc.

### bokeh: 
> Paquete de visualización, capaz de crear graficos estadisticos interactivos, por medio del uso de jasvascript y de un navegador web.

### folium: 
> Paquete que permite realizar gráficos y visualizaciones estadísticas, que tengan relación con mapas o información geográfica.
